```python
# Load paste.txt, rename columns appropriately and create chart
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import io

# Correctly parse the data by reading the file line by line
with open('paste.txt', 'r') as f:
    lines = f.readlines()

# The header is the first line, split by '|' and then by ',' and strip whitespace
header_parts = [h.strip() for h in lines[0].split('|')]
header = []
for part in header_parts:
    header.extend([h.strip() for h in part.split(',')])


# The data starts from the second line, split by '|' and then by ','
data = []
for line in lines[1:]:
    parts = [p.strip() for p in line.split('|')]
    date_part = parts[0]
    qty_price_parts = parts[1].split(',')
    data.append([date_part, qty_price_parts[0], qty_price_parts[1]])

# Create a DataFrame
paste_df = pd.DataFrame(data, columns=header)

print(paste_df.head())

# Clean column names
paste_df.columns = [c.strip() for c in paste_df.columns]

# Rename expected columns
col_date = 'Business Date'
col_qty = 'On Loan Quantity'
col_price = 'Security Price'

# Convert columns to appropriate data types
paste_df[col_date] = pd.to_datetime(paste_df[col_date])
paste_df[col_qty] = pd.to_numeric(paste_df[col_qty])
paste_df[col_price] = pd.to_numeric(paste_df[col_price])


paste_df = paste_df.sort_values(col_date)

fig, ax1 = plt.subplots(figsize=(12,7))

border_color = '#048dd2'
loan_color = '#f57600'
fig.patch.set_edgecolor(border_color)
fig.patch.set_linewidth(16)

ax1.plot(paste_df[col_date], paste_df[col_price], color=border_color, linewidth=3.0, label='Price per Share (USD)')
ax1.set_ylabel('Price per Share (USD)', fontsize=12, fontweight='bold', color='#374151', labelpad=8)
ax1.tick_params(axis='y', labelcolor=border_color, labelsize=10)
for lab in ax1.get_yticklabels(): lab.set_fontweight('bold')
ax1.set_xlim(left=paste_df[col_date].min())
ax1.set_xmargin(0)

ax2 = ax1.twinx()
ax2.plot(paste_df[col_date], paste_df[col_qty]/1e6, color=loan_color, linewidth=2.5, label='Loan Quantity (MM shares)')
ax2.set_ylabel('Loan Quantity (MM shares)', fontsize=12, fontweight='bold', color='#374151', rotation=270, labelpad=15, va='bottom')
ax2.tick_params(axis='y', labelcolor=loan_color, labelsize=10)
ax2.yaxis.set_major_formatter(FuncFormatter(lambda x,pos: format(x,'.2f')))
for lab in ax2.get_yticklabels(): lab.set_fontweight('bold')

plt.title('Default Chart', fontsize=16, fontweight='bold', color='#006db7')

lines, labels = [], []
for ax in [ax1, ax2]:
    lns, lbl = ax.get_legend_handles_labels()
    lines += lns
    labels += lbl
leg = plt.legend(lines, labels, loc='upper left', frameon=True, fontsize=12)
leg.get_frame().set_facecolor('white'); leg.get_frame().set_edgecolor('none')
for txt in leg.get_texts(): txt.set_fontweight('bold')

for spine in ax1.spines.values(): spine.set_edgecolor('#BDDFFF'); spine.set_linewidth(1.5)
for spine in ax2.spines.values(): spine.set_edgecolor('#BDDFFF'); spine.set_linewidth(1.5)

plt.annotate('Source: EquiLend Data & Analytics', (0.975, 0.025), xycoords='figure fraction', ha='right', fontsize=12, fontweight='bold', color='#006db7')
plt.tight_layout()
plt.show()
```

## Load data

### Subtask:
Load the datasets from `/content/kohls_kss.csv` and `/content/opendoor.csv` into pandas DataFrames.


**Reasoning**:
The subtask requires loading two CSV files into pandas DataFrames. I will use `pd.read_csv` for this purpose and display the head of both dataframes to verify the loading.



```python
kohls_df = pd.read_csv('/content/kohls_kss.csv')
opendoor_df = pd.read_csv('/content/opendoor.csv')

print("Kohls DataFrame head:")
display(kohls_df.head())

print("\nOpendoor DataFrame head:")
display(opendoor_df.head())
```

## Explore data

### Subtask:
Display the first few rows and column information for each DataFrame to understand their structure and content.


**Reasoning**:
Display the head and info for both dataframes as requested in the instructions.



```python
print("Kohls DataFrame head:")
display(kohls_df.head())
print("\nKohls DataFrame info:")
kohls_df.info()

print("\nOpendoor DataFrame head:")
display(opendoor_df.head())
print("\nOpendoor DataFrame info:")
opendoor_df.info()
```

## Merge data

### Subtask:
Merge the two DataFrames based on a common column, likely a date column, to combine the information for analysis.


**Reasoning**:
Convert the 'Business Date' column to datetime objects in both dataframes and then merge them using an inner join on the 'Business Date' column. Finally, display the head of the merged dataframe.



```python
kohls_df['Business Date'] = pd.to_datetime(kohls_df['Business Date'])
opendoor_df['Business Date'] = pd.to_datetime(opendoor_df['Business Date'])

merged_df = pd.merge(kohls_df, opendoor_df, on='Business Date', how='inner')

print("Merged DataFrame head:")
display(merged_df.head())
```

## Analyze data

### Subtask:
Perform the requested analysis or visualization on the merged dataset. Since no specific analysis was requested, I will perform a basic analysis like plotting the closing prices of the stocks over time.


**Reasoning**:
Create a line plot to visualize the 'Security Price' over time for both Kohls and Opendoor from the merged dataframe.



```python
plt.figure(figsize=(12, 7))
plt.plot(merged_df['Business Date'], merged_df['Security Price_x'], label='Kohls Security Price')
plt.plot(merged_df['Business Date'], merged_df['Security Price_y'], label='Opendoor Security Price')
plt.xlabel('Date')
plt.ylabel('Security Price')
plt.title('Security Price Comparison: Kohls vs. Opendoor')
plt.legend()
plt.grid(True)
plt.show()
```
